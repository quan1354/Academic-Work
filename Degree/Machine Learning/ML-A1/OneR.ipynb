```python
import pandas as pd
import numpy as np
df=pd.read_excel('Freq-PHO-Binary.csv.xlsx')
print(df)
list = df.values.tolist()
result_list = df['Depression'].tolist()
df2=df.transpose()
df_numeric=df.drop(['Gender','Depression'],axis=1)
list2 = df2.values.tolist()
df_numeric=df_numeric.transpose()
numeric_list=df_numeric.values.tolist()

#calculate gender feature confusion matrix seperately
count={}
count_other=[]
for i in range(len(list2)):
  for j in range(len(list2[i])):
    count[j]=0

count_gender=[[0 for i in range(2)]for j in range(2)]
print("size:")
print(len(count_gender))
print(len(count_gender[0]))

for i in range(len(list)):
  for j in range(len(list[i])):
    if result_list[i]=='YES':
      count[j]=count[j]+1

print(count)

count_result=0
for i in range(len(result_list)):
  if (result_list[i]=='YES'):
    count_result=count_result+1
print("YES count: "+str(count_result))
for i in range(len(list2)):
  for j in range(len(list2[i])):
    #gender result
    if i==0:
      if result_list[j]=='YES':
        #print('test')
        if (list2[i][j]=='Male'):
          #print('test2')
          count_gender[0][0]=count_gender[0][0]+1
        else:
          count_gender[0][1]=count_gender[0][1]+1
      else:
        if (list2[i][j]=='Male'):
          count_gender[1][0]=count_gender[1][0]+1
        else:
          count_gender[1][1]=count_gender[1][1]+1
gender_matrix=[[0 for i in range(2)]for j in range(2)]

gender_predict=[True,True]

#produce prediction
if count_gender[1][0]>count_gender[0][0]:
  gender_predict[0]=False

if count_gender[1][1]>count_gender[0][1]:
  gender_predict[1]=False


#calculate confusion matrix
for j in range(len(list2[0])):
    #gender result
    if result_list[j]=='YES':
      if (list2[0][j]=='Male'):
        #True positive
        if gender_predict[0]==True:
            #print('test2')
          gender_matrix[0][0]+=1
        #False negative
        else:
          gender_matrix[0][1]+=1
      else:
        #True positive
        if gender_predict[1]==True:
          gender_matrix[0][0]+=1
        #False negative
        else:
          gender_matrix[0][1]+=1
    else:
      if (list2[0][j]=='Male'):
        #False positive
        if gender_predict[0]==True:
          gender_matrix[1][0]+=1
        #True negative
        else:
          gender_matrix[1][1]+=1
      else:
        #False positive
        if gender_predict[1]==True:
          gender_matrix[1][0]+=1
        #True negative
        else:
          gender_matrix[1][1]+=1

truePN=gender_matrix[0][0]+gender_matrix[1][1]
print(truePN)
gender_accuracy=truePN/np.sum(gender_matrix)
          
        
print(np.matrix(numeric_list))
print(np.matrix(gender_matrix))
print("Gender attribute accuracy: "+str(gender_accuracy*100)+"%")

#slice the numbers into a fixed range numbers

new_numeric_list=numeric_list
temp=[]

print(np.matrix(new_numeric_list))
for i in range(len(numeric_list)):
  temp=numeric_list[i]
  print(str(i)+":")
  tempMax=max(temp)
  print(tempMax)
  tempMin=min(temp)
  tempRange=(tempMax-tempMin)/3 
  print(tempRange)
  tempRange=(tempMax-tempMin)/3
  for j in range(len(numeric_list[0])):
    
    if numeric_list[i][j] <= tempRange:
      new_numeric_list[i][j]=1
    elif numeric_list[i][j] <= tempRange*2:
      new_numeric_list[i][j]=2
    elif numeric_list[i][j] <= tempMax:
      new_numeric_list[i][j]=3


print(np.matrix(new_numeric_list))  

predict_list=[[0 for i in range(len(new_numeric_list))]for j in range(3)]
print(np.matrix(predict_list))
#decide yes or no prediction
for i in range(len(new_numeric_list)):
  temp_yes=[0,0,0]
  temp_no=[0,0,0]
  for j in range(len(new_numeric_list[0])):
    if (result_list[j]=="YES"):
      if(new_numeric_list[i][j]==1):
        temp_yes[0]+=1
      elif(new_numeric_list[i][j]==2):
        temp_yes[1]+=1
      elif(new_numeric_list[i][j]==3):
        temp_yes[2]+=1
    else:
      if(new_numeric_list[i][j]==1):
        temp_no[0]+=1
      elif(new_numeric_list[i][j]==2):
        temp_no[1]+=1
      elif(new_numeric_list[i][j]==3):
        temp_no[2]+=1
  for count in range (len(predict_list)):
    if(temp_yes[count]>=temp_no[count]):
      predict_list[count][i]=True
    else:
      predict_list[count][i]=False
#produce matrix
shape=(len(new_numeric_list),2,2)
number_matrix=np.zeros(shape)
print((number_matrix))

for i in range(len(new_numeric_list)):
  for j in range(len(new_numeric_list[0])):
    if (result_list[j]=="YES"):
      if(new_numeric_list[i][j]==1):
        #true positive
        if (predict_list[0][i]==True):
          number_matrix[i][0][0]+=1
        #false negative
        else:
          number_matrix[i][0][1]+=1
      elif(new_numeric_list[i][j]==2):
        #true positive
        if predict_list[1][i]==True:
          number_matrix[i][0][0]+=1
        else:
          number_matrix[i][0][1]+=1
          
      elif(new_numeric_list[i][j]==3):
        #true positive
        if predict_list[2][i]==True:
          number_matrix[i][0][0]+=1
        else:
          number_matrix[i][0][1]+=1
    else:
      if(new_numeric_list[i][j]==1):
        #False positive
        if predict_list[0][i]==True:
          number_matrix[i][1][0]+=1
        #True negative
        else:
          number_matrix[i][1][1]+=1
      elif(new_numeric_list[i][j]==2):
        #False positive
        if predict_list[1][i]==True:
          number_matrix[i][1][0]+=1
        else:
          number_matrix[i][1][1]+=1
      elif(new_numeric_list[i][j]==3):
        #False positive
        if predict_list[2][i]==True:
          number_matrix[i][1][0]+=1
        else:
          number_matrix[i][1][1]+=1

numeric_accuracy=[0 for i in range(len(new_numeric_list))]

print(number_matrix)

#compute accuracy
for i in range(len(number_matrix)):
  print(len(number_matrix))
  sum=0
  for j in range(len(number_matrix[0])):
    for a in range(len(number_matrix[0][0])):
      sum+=number_matrix[i][j][a]
  print("sum= "+str(sum))
  numeric_accuracy[i]=(number_matrix[i][0][0]+number_matrix[i][1][1])/(sum)

print(np.matrix(numeric_accuracy))

```

```python
import urllib.request

url_train = "http://fashion-mnist.s3-website.eu-central-1.amazonaws.com/train-images-idx3-ubyte.gz"
url_train_labels = "http://fashion-mnist.s3-website.eu-central-1.amazonaws.com/train-labels-idx1-ubyte.gz"
url_test = "http://fashion-mnist.s3-website.eu-central-1.amazonaws.com/t10k-images-idx3-ubyte.gz"
url_test_labels = "http://fashion-mnist.s3-website.eu-central-1.amazonaws.com/t10k-labels-idx1-ubyte.gz"

# Download the dataset
urllib.request.urlretrieve(url_train, "train-images-idx3-ubyte.gz")
urllib.request.urlretrieve(url_train_labels, "train-labels-idx1-ubyte.gz")
urllib.request.urlretrieve(url_test, "t10k-images-idx3-ubyte.gz")
urllib.request.urlretrieve(url_test_labels, "t10k-labels-idx1-ubyte.gz")
```

```python
import tensorflow as tf

# Load the dataset
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.fashion_mnist.load_data()

# Print shapes 
print("Train images:", train_images.shape)
print("Train labels:", train_labels.shape)
print("Test images:", test_images.shape)
print("Test labels:", test_labels.shape)
```

```python
import matplotlib.pyplot as plt
import numpy as np

class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

# Display some images from the dataset
fig, axes = plt.subplots(nrows=5, ncols=5, figsize=(10,10))
axes = axes.ravel()

for i in np.arange(0, 25):
  index = np.random.randint(0, len(train_images))
  axes[i].imshow(train_images[index], cmap='gray')
  axes[i].set_title(class_names[train_labels[index]])
  axes[i].axis('off')
plt.subplots_adjust(hspace=0.5)
plt.show()

print("train_images[1]:")
print(train_images[1])
print("")
print("train_labels:")
print(train_labels)
print("test_labels:")
print(test_labels)
```

```python
#Normalisation
train_images = train_images / 255.0
test_images = test_images / 255.0
train_images = train_images.reshape(train_images.shape[0], -1)
test_images = test_images.reshape(test_images.shape[0], -1)
print("train_images[1]:")
print(train_images[1])
print(test_images[1])

#One Hot Encoding
train_labels = train_labels.astype(int)
train_labels = np.eye(10)[train_labels]
test_labels = test_labels.astype(int)
test_labels = np.eye(10)[test_labels]
print (train_labels[1])
print(test_labels[1])
```

```python
import numpy as np
import matplotlib.pyplot as plt

input_size = 784
output_size = 10
learning_rate = 0.001
batch_size = 20
batches = train_images.shape[0] // batch_size
train_loss_data = []
train_acc_data = []
test_loss_data = []
test_acc_data = []
q = 0
w1 = np.random.randn(input_size, output_size) * 0.01
b1 = np.zeros((1, output_size))


def sigmoid(x):
  x = np.clip(x, -500, 500)
  return(1/(1 + np.exp(-x)))

def sigmoid_deriv(x):
  return sigmoid(x) * (1 - sigmoid(x))

def softmax_deriv(x):
  exp_x = np.exp(x)
  return exp_x / np.sum(exp_x, axis=1, keepdims=True)*(1 - exp_x / np.sum(exp_x, axis=1, keepdims=True))

def softmax(x):
  exponent = np.exp(x.astype(np.float128))
  return exponent/exponent.sum(axis=1,keepdims=True)

def feed_forward(x, w1, b1):
  # output
  z1 = np.dot(x, w1)+b1
  output_out = softmax(z1)
  return(output_out)

def back_prop(train_label, output_out, batch_images, w1, b1, learning_rate):
  error = output_out - train_label
  dW = np.dot(batch_images.T, error)
  db = np.sum(error, axis=0, keepdims=True)
 
  w1 -= learning_rate *dW
  b1 -= learning_rate *db
  return(w1, b1)

def compute_loss(y_pred, y_true):
  n = len(y_pred)*2
  loss = np.sum(np.square(y_pred - y_true))/n
  return loss

def accuracy(y_pred, y_true):
  m = y_pred.shape[0]
  n_classes = y_pred.shape[1]
  # One-hot encoding
  Y_pred_onehot = np.zeros_like(y_pred)
  Y_pred_onehot[np.arange(m), y_pred.argmax(axis=1)] = 1
  # Correct predictions
  num_correct = np.sum(Y_pred_onehot * y_true)
  acc = (num_correct / m)*100
  return acc

for epoch in range (10):
  for i in range(batches):
    batch_images = train_images[i*batch_size:(i+1)*batch_size]
    batch_labels = train_labels[i*batch_size:(i+1)*batch_size]
    a1 = feed_forward(batch_images,w1,b1)
    w1, b1= back_prop(batch_labels,a1,batch_images,w1,b1,learning_rate)
    
  xa1 = feed_forward(train_images,w1,b1)
  train_loss = compute_loss(xa1, train_labels)
  train_loss_data.append(train_loss)
  train_acc = accuracy(xa1, train_labels)
  train_acc_data.append(train_acc)
  print("Training...")
  print(f"Epoch: {epoch}, Loss: {train_loss:.4f}, Accuracy: {train_acc:.2f}%")
  if((train_loss > train_loss_data[epoch-1])and(train_acc< train_acc_data[epoch-1])):
    break
  
#With test data
predict_a1= feed_forward(test_images,w1,b1)
predict_loss = compute_loss(predict_a1, test_labels)
test_loss_data.append(predict_loss)
predict_acc = accuracy(predict_a1, test_labels)
test_acc_data.append(predict_acc)
print("")
print("Predict")
print(f"Loss: {predict_loss:.4f}. Accuracy: {predict_acc:.2f}%")
    
plt.plot(train_loss_data)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title("Training Loss")
plt.show()

plt.plot(train_acc_data)
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title("Training Accuracy")
plt.show()
```
